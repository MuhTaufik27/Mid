```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

# Baca dataset
file_path = "/content/nvda_data.csv"  # Sesuaikan path jika diunggah ke Colab
data = pd.read_csv(file_path)

# Tampilkan 5 baris pertama
print("Sample Data:")
print(data.head())
```

```python
# Ukuran dataset
print("\nJumlah Baris dan Kolom:")
print(data.shape)

# Tipe data tiap kolom
print("\nTipe Data:")
print(data.dtypes)

# Jumlah nilai unik untuk setiap kolom kategori
print("\nJumlah Nilai Unik untuk Setiap Kolom Kategori:")
for col in data.select_dtypes(include=['object', 'category']).columns:
    print(f"{col}: {data[col].nunique()} unique values")
```

```python
# Statistik deskriptif kolom numerik
print("\nStatistik Deskriptif Kolom Numerik:")
print(data.describe())

# Identifikasi outliers menggunakan metode IQR
def detect_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return col[(col < lower_bound) | (col > upper_bound)]

print("\nOutliers di Kolom Numerik:")
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    outliers = detect_outliers(data[col])
    print(f"{col}: {len(outliers)} outliers")
```

```python
# Cek nilai kosong
print("\nMissing Values:")
missing = data.isnull().sum()
missing_percentage = (missing / len(data)) * 100
missing_data = pd.DataFrame({'Missing Values': missing, 'Percentage': missing_percentage})
print(missing_data[missing_data['Missing Values'] > 0])

# Rencana penanganan missing values
# Contoh: Mengisi dengan mean untuk numerik, modus untuk kategori
data.fillna(data.mean(numeric_only=True), inplace=True)  # Isi nilai numerik dengan rata-rata
data.fillna(data.mode().iloc[0], inplace=True)          # Isi nilai kategori dengan modus
```

```python
# Histogram untuk kolom numerik
data.select_dtypes(include=['float64', 'int64']).hist(bins=20, figsize=(15, 10))
plt.suptitle('Histogram Kolom Numerik')
plt.show()

# Uji normalitas (Shapiro-Wilk Test)
print("\nUji Normalitas (Shapiro-Wilk Test):")
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    stat, p = shapiro(data[col].dropna())
    print(f"{col}: p-value = {p:.5f} {'(Tidak Normal)' if p < 0.05 else '(Normal)'}")
```

```python
# Korelasi antar variabel numerik
# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_data.corr()

print("\nMatriks Korelasi:")
print(correlation_matrix)

# Variabel dengan korelasi tertinggi
strong_corr = correlation_matrix.unstack().sort_values(ascending=False)
strong_corr = strong_corr[strong_corr < 1]  # Hilangkan korelasi dengan diri sendiri
print("\nPasangan Variabel dengan Korelasi Tertinggi:")
print(strong_corr.head())

# Scatter plot untuk variabel dengan korelasi tertinggi
var1, var2 = strong_corr.index[0]
sns.scatterplot(data=data, x=var1, y=var2)  # Use original data for plotting
plt.title(f"Scatter Plot: {var1} vs {var2}")
plt.show()
```

```python
# Box plot untuk satu kolom numerik
numerical_col = 'Open'  # Ganti dengan nama kolom numerik yang ada di DataFrame, contoh: 'Open'
sns.boxplot(data=data, y=numerical_col)
plt.title(f"Box Plot: {numerical_col}")
plt.show()

# Jumlah outliers
outliers = detect_outliers(data[numerical_col])
print(f"\nJumlah Outliers di {numerical_col}: {len(outliers)}")
```

```python
# Import library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt

# Baca dataset
file_path = "/content/nvda_data.csv"  # Sesuaikan path file jika diperlukan
data = pd.read_csv(file_path)

# Pastikan kolom waktu ada dan dalam format datetime
# Ganti 'Date' dengan 'Datetime', yang merupakan nama kolom waktu di dataset Anda
data['Datetime'] = pd.to_datetime(data['Datetime'])

# Filter data untuk tahun 2024
data_2024 = data[data['Datetime'].dt.year == 2024]

# Cek data tahun 2024
print("Data Tahun 2024:")
print(data_2024.head())
```

```python
# Tambahkan kolom bulan untuk agregasi
data_2024['Month'] = data_2024['Datetime'].dt.month

# Hitung rata-rata harga saham per bulan
monthly_avg = data_2024.groupby('Month')['Close'].mean()  # Ganti 'Close' dengan kolom harga penutupan saham Anda

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_avg.index, monthly_avg.values, marker='o', color='blue')
plt.title('Rata-rata Harga Saham per Bulan di Tahun 2024')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Harga Saham (Close)')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Cari harga terendah di tahun 2024 (waktu terbaik membeli saham)
buy_date = data_2024.loc[data_2024['Close'].idxmin(), 'Datetime']  # Changed 'Date' to 'Datetime'
buy_price = data_2024['Close'].min()

# Cari harga tertinggi setelah pembelian (waktu terbaik menjual saham)
data_after_buy = data_2024[data_2024['Datetime'] > buy_date]  # Changed 'Date' to 'Datetime'
sell_date = data_after_buy.loc[data_after_buy['Close'].idxmax(), 'Datetime']  # Changed 'Date' to 'Datetime'
sell_price = data_after_buy['Close'].max()

# Cetak hasil
print(f"Waktu terbaik untuk membeli saham: {buy_date.strftime('%d %B %Y')} dengan harga {buy_price}")
print(f"Waktu terbaik untuk menjual saham: {sell_date.strftime('%d %B %Y')} dengan harga {sell_price}")
```
